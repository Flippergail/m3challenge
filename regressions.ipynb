```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) * 10
y = 2 * X + 1 + np.random.randn(100, 1) * 2
```

```python
reg = np.polyfit(X.flatten(), y.flatten(), 1)
```

```python
trend = np.polyval(reg, X.flatten())
plt.plot(X, y, 'b.')
plt.plot(X, trend, 'r-')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
